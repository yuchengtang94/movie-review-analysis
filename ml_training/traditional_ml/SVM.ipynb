```python
import csv
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
from sklearn.model_selection import train_test_split   
from sklearn.svm import SVC
```

```python
train = pd.read_csv('../input/train.tsv', sep='\t')
test = pd.read_csv('../input/test.tsv',  sep='\t')
sampleSub = pd.read_csv('../input/sampleSubmission.csv')
```

```python
x_train, x_test, y_train, y_test = train_test_split(train.Phrase, train.Sentiment, test_size = 0.2)
```

```python
cv = CountVectorizer(max_features = None)
```

```python
cv.fit(x_train)
```

```python
x_train_cv = cv.transform(x_train)
x_test_cv = cv.transform(x_test)
```

```python
SVM_classifier = SVC()
```

```python
SVM_classifier.fit(x_train_cv, y_train)
```

```python
y_pred = SVM_classifier.predict(x_test_cv)
print('Accuracy of test: ', metrics.accuracy_score( y_pred , y_test))
```

```python
x_final_test = test.Phrase
x_final_test_cv = cv.transform(x_final_test)
y_final_test_pred = SVM_classifier.predict(x_final_test_cv)
sampleSub.Sentiment = y_final_test_pred
sampleSub.to_csv("naive_bayes_cv.csv", index=False)

from joblib import dump, load
nb_path = 'model/svm_cv.joblib'
dump(SVM_classifier, nb_path)
```
